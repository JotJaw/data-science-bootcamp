```python
import numpy as np
np.__version__
```

```python
price = np.array([[12.40, 12.80, 11.90, 12.60, 1000],
                  [12.50, 13.00, 11.70, 12.20, 2000],
                  [12.20, 13.40, 12.20, 13.20, 1500]])
print(price)
```

```python
price.sum() #suma wszystkich elementów w tablicy
```

```python
#suma w każdym wierszu lub kolumnie - parametr, po której osi liczyć 0 - suma poszczególnych kolumn, 1 - wierszy
price.sum(axis=0)
```

```python
price.sum(axis=1)
```

```python
np.sum(price, axis=0) #inny sposób na to samo
```

```python
np.min(price, axis=0)
```

```python
np.max(price)
```

```python
np.median(price, axis=0)
```

```python
np.mean(price, axis = 0)
```

```python
np.std(price, axis=0)
```

```python
np.set_printoptions(precision=4,suppress=True) #ustawienie dla printów - supress True - bez notacji matematycznej
```

```python

np.var(price, axis = 0)
```
