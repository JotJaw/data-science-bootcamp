# Maski logiczne

```python
A = np.arange(-10,10,0.5)
A = A.reshape(10,-1)
A
```

```python
#gdzie mamy wartości ujemne?
A > 0
```

```python
#dzięki masce możemy wyciąć interesujące nas elementy
A[A > 0]
```

```python
#łączenie warunków
# A[A > -5 and A < 5] tak się nie da
np.bitwise_and(A > -5, A < 5) #zwraca maskę, dzięki której możemy wybrać interesujące nas elementy
```

```python
A[np.bitwise_and(A > -5, A < 5)]
```

```python
np.bitwise_or(A < -5, A > 5)
```
