```python
import matplotlib
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

#Wykres liniowy

```python
_ = plt.plot([0,1],[0,3])
```

```python
plt.plot([0,1,2],[0,3,2], label='line', color='green') #domyślnie wykres liniowy
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres liniowy')
_ = plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
plt.xlabel('oś x')
plt.ylabel('oś y')
_ = plt.title('Wykres sin(x) i cos(x)')
```

#Tworzenie wielu wykresów

```python
plt.subplot(121) #1 - liczba wierszy w siatce, 2 - liczba kolumn w siatce, 1 - wykres pierwszy po lewej
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,1,4], 'r--') #czerwona linia przeywana
```

```python
fig = plt.figure(figsize=(12,8)) #domyślnie przy wykonuwaniu kodu w pojednyczej komórce obiekt figure automatycznie utworzony, ale można ręcznie - jak tutaj - dzięki temu jest możliwość kontroli parametrów

plt.subplot(131) #1 - liczba wierszy w siatce, 3 - liczba kolumn w siatce (tu: 3 wykresy obok siebie), 1 - wykres pierwszy po lewej
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--') #czerwona linia przeywana
plt.subplot(133)
plt.plot([0,1,2],[3,1,4], 'ro')

#w pycharm konieczna jest metoda fig.show() w notebookach nie trzeba, obraz jest domyślnie wyświetlany
```

#Wykresy słupkowe

```python
plt.bar(x=[0,1,2,3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2, 3.2, 1.2, 5.2]) #wykres horyzontalny
```

#Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2) #prosty wykres rozrzutu
```

```python
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) #alpha - przezroczystość punktów
```
