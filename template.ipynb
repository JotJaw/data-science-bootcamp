Typy danych

```python
#do array można przekazać dodatkowy parametr - dtype, aby kontrolować typ
A = np.array([1, 2.3, 3.3], dtype='float')
A.dtype
```

```python
#liczby złożone
A = np.array([1,2,3], dtype='complex')
A.dtype
```

```python
A = np.array([1, 2.3, 3.3], dtype='int') #konwersja na liczby całkowite
A.dtype
```

```python
A
```

```python
A = np.array([True, False])
A.dtype
```

```python
A = np.array([24,120,230], dtype=np.int16)
A.dtype
```

```python
#unsigned int - do 255
A = np.array([24,120,230], dtype=np.uint8)
A.dtype
```
