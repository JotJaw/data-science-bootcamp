##Wycinanie

```python
price = pd.Series(data={'Apple':200,'CD Projekt':60,'Amazon':1900,'KGHM':np.nan}) #index -> klucz ze słownika
price
```

```python
price['CD Projekt']
```

```python
price[1]
```

##Podstawowe funkcje

```python
price.count() #ile elementów w obiekcie - zlicza tylko dane występujące
```

```python
price.value_counts() #rozkład danych - ile razy dana wartość; użyteczna do budowy histogramu
```

nie zlicza braku, chyba że ustawimy paramet dropna=False

```python
price.value_counts(dropna=False)
```

```python
price.sum()
```

```python
price.max()
```

```python
price.min()
```

```python
price.std()
```

```python
price.describe() #kwantyle rozkładu (50% = median)
```

```python
price.nlargest(2) #największe 2
```

```python
price.nsmallest(2) #najmniejsze - jako parametr - ile
```

```python
price.rank() #uporządkowanie od najniższej - przypisanie wartości
```

```python
price.sort_values() #braki danych na końcu
```

```python
price.sort_values(ascending=False) #zmiana porządku sortowania
```

###metoda apply()

```python
price_pln = price.apply(lambda x: x * 3.8) #dane w PLN; funkcja może być osobno zdefiniowana i przekazana jako parametr; jeżeli prosta - może być zdefiniowana przy pomocy instrukcji lamda
```

```python
price #bez zmian - nie zostały przypisane do obiektu
```

```python
price_pln
```
