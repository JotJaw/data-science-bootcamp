#Podstawowe struktury danych: pd.Series

```python
s = pd.Series(data=[1,2,4,6])
s
```

indeks można dowolnie zmienić

za indeks można użyć daty!

```python
s = pd.Series(data=[1.,2,4,6], index=['a','b','c','d'], name ='sample') #jeden float, wszędzie float
s
```

wstawianie braków danych

```python
import numpy as np
np.nan
```

brak wartości - wartość NaN nie zostanie wzięta pod uwagę przy wyliczeniach - w przeciwieństwie do np. wartości 0

```python
s = pd.Series(data=[1.,np.nan,4,6], index=['a','b','c','d'], name ='sample') #jeden float, wszędzie float
s
```
