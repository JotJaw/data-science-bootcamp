wykresów -> ułatwiona manipulacja danychmi i generowanie wykresów na podstawie istniejących zbiorów danych

* wykresy statystyczne - punktowe, box plots, wykresy rozkładu, wiolinowe, słupkowe -> wizualizacja rozkładów danych i relacji między danymi

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.__version__
```

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
df.info() #category - typ danych, w których można określić konkretną kategorię (słownik?)
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip')
```

```python
sns.set(font_scale=1.2)
_ = sns.relplot(data=df, x='total_bill', y='tip')
```

```python
sns.set(font_scale=1.2)
_ = sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue = 'size')
#kolor - hue - tu w zależności od size
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue = 'size', col='time')
#col - kolumna, dzięki której dane zostaną podzielone, najlepiej kategoryczne o niewielkiej liczbie wartości
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue = 'size', row='time')
#row - dane zostaną podzielone w pionie
```

```python
_ = sns.relplot(data=df, x='total_bill', y='tip', size = 'size', hue = 'size', col='time', row='smoker')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='swarm')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='box') #rozkład zmiennej total_bill w rozbiciu na każdy dzień
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='violin')
```

```python
_ = sns.catplot(data=df, x='day', y='total_bill', kind='bar')
```

```python
_ = sns.catplot(data=df, x='day', kind='count') #częstość występowania każdego dnia (częstotliwość występowanie danej kategorii)
```
