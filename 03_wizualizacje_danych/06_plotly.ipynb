#Plotly

```python
import plotly
import plotly.graph_objects as go
plotly.__version__
```

##Wykres słupkowy

```python
fig = go.Figure(
    data=[go.Bar(y=[2, 1, 3, 4])], #jeden element "normalnie", więcej - jako listę
    #layout_title_text="A Figure Displayed with fig.show()"
)
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 1, 3, 4]),
    layout=go.Layout(title={'text': 'Wykres słupkowy'})
    )
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2, 1, 3, 4]),
    layout=go.Layout(title_text='Wykres słupkowy')
    )
#zapis do pliku html - można pobrać i z nim pracować
fig.write_html('bar.html')
fig.show()
```

##Wykres kołowy i pierścieniowy -Diamonds

```python
from seaborn import load_dataset
df = load_dataset('diamonds')
df.head()
```

```python
df.info()
```

```python
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns = ['type', 'count']
dff
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count']),
    layout=go.Layout(title={'text': 'Rozkład zmiennej cut'}, width=800)
)
fig.show()
```

```python
#wykres pierścieniowy - z parametrem hole
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'], hole = 0.5),
    layout=go.Layout(title={'text': 'Rozkład zmiennej cut'}, width=800)
)
fig.show()
```

##Diagram Sankey'a

np. przepływ klientów

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019']),
                 link=dict(source=[0, 0, 1, 1], # indeks odpowiadający etykiecie (labels)
                          target=[2, 3, 2, 3],
                          value=[65, 12, 18, 5]))] #wartość przepuszczana od źródła do celu

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
#można przesuwać po wyświetleniu
```

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019',
                                   'Churn_2019', 'Nonchurn_2020', 'Churn_2020']),
                 link=dict(source=[0, 0, 1, 1, 2, 3], # indeks odpowiadający etykiecie (labels)
                          target=[2, 3, 2, 3, 4, 5],
                          value=[65, 10, 5, 20, 70, 30]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019',
                                   'Churn_2019', 'Nonchurn_2020', 'Churn_2020']),
                 link=dict(source=[0, 0, 1, 1, 2, 2, 3, 3], # indeks odpowiadający etykiecie (labels)
                          target=[2, 3, 2, 3, 4, 5, 4, 5],
                          value=[65, 10, 5, 20, 60, 10, 15, 15]))]

fig = go.Figure(data=data, layout=go.Layout(width=800, height=400))
fig.show()
```

##Wykres świecowy

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df = df[df.Date > '2026-01-01']
df.head(3)
```

```python
fig = go.Figure(data=go.Candlestick(x=df.Date, open=df.Open, high=df.High, low=df.Low, close=df.Close))
fig.show()
```
