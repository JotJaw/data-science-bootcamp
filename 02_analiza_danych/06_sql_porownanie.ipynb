### <a name='a2'></a>Wczytanie danych i preprocessing

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >= 0] #nadpisanie częścią rekordów
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x))) #zmiana typu kolumny z float na string z inta
retail.info()
```

```python
retail.head()
```

### <a name='a3'></a>Porównanie do języka SQL

```python
# select * from retail
retail
```

```python
# select Quantity, UnitPrice, CustomerId from retail
retail[['Quantity', 'UnitPrice', 'CustomerID']]
```

```python
# SELECT Quantity, UnitPrice, CustomerID FROM retail LIMIT 10; /*10 pierwszych wierszy*/
retail[['Quantity', 'UnitPrice', 'CustomerID']].head(10)
```

```python
retail[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```python
# SELECT * FROM retial WHERE CustomerID='17850';
retail.CustomerID == '17850'
```

```python
retail[retail.CustomerID == '17850'] #wykorzystujemy maskowanie logiczne
```

```python
retail.query('CustomerID == "17850"') #metoda query
```

```python
retail.query("CustomerID == '17850'") #metoda query; może być też tak - najpierw podwójny cudzysłów, potem ciapki
```

```python
# SELECT * FROM retial WHERE CustomerID='17850' and UnitPrice > 5;
retail[(retail.CustomerID == '17850') & (retail.UnitPrice > 5)]
```

```python
# SELECT * FROM retial WHERE CustomerID='17850' or Country ='France';
retail[(retail.CustomerID == '17850') | (retail.Country == 'France')]
```

```python
#select * from retail where InvoiceNo is not null;
retail[retail.InvoiceNo.notnull()]
```

```python
retail[retail.InvoiceNo.isnull()]
```
