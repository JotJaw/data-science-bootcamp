# 4. Odległość dwóch punktów w przestrzeni $R^n$
\begin{aligned}
d(\mathbf {q}, \mathbf {p}) = d(\mathbf {p}, \mathbf {q})
&= \sqrt{{(q_1 - p_1)}^2+{(q_2 - p_2)}^2+ ... + {(q_n - p_n)}^2}\\
&= \sqrt{\sum_{i=1}^n{(q_i - p_i)}^2}
\end{aligned}


liczymy normę wekora, który powstaje - pierwiastek sumy kwadratów różnic
