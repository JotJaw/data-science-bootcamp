### <a name='a2'></a>Wczytanie danych i preprocessing

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >= 0] #nadpisanie częścią rekordów
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x))) #zmiana typu kolumny z float na string z inta
retail.info()
```

```python
retail.head()
```

####Grupowanie

```python
# SELECT CustomerID, count(*) FROM retial GROUP BY CustomerID;
retail.groupby('CustomerID').size()
```

```python
retail.head()
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;
#przychód na pojedyńczą transakcję
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail.head()
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})
#słownik z mapowaniem nazw kolumn na funkcje, których chcemy użyć
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': 'mean', 'CustomerID': np.size})
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': "mean", 'CustomerID': "size"}). \
        rename(columns={'Revenue': 'RevenueAverage', 'CustomerID': 'CustomerIDCount'})
```

```python
retail.nsmallest(n=5, columns='Quantity').astype({'StockCode': 'str'}).sort_values(by='StockCode')
```

```python
results = retail.groupby('InvoiceDate').aggregate({'Revenue': 'sum'}).rename(columns={'Revenue': 'RevenueSum'})
results
```

```python
import seaborn as sns
sns.set()

results.plot(kind='bar', figsize=(12, 5))

#
```

####Sortowanie

```python
# SELECT * FROM retail ORDER BY Quantity DESC LIMIT 5;
retail.nlargest(5, columns='Quantity')
```

```python
# SELECT * FROM retail ORDER BY Quantity LIMIT 5;
retail.nsmallest(n=5, columns='Quantity')
```

retail.nsmallest(n=5, columns='Quantity').sort_values(by='StockCode')

TypeError: '<' not supported between instances of 'str' and 'int'

The error is due to a TypeError when trying to sort the StockCode column. This column contains a mix of strings and numbers, which Python cannot directly compare. To resolve this, we need to ensure the entire StockCode column is treated as strings before sorting. I will update the code to convert the 'StockCode' column to string type.

```python
retail.nsmallest(n=5, columns='Quantity').astype({'StockCode': 'str'}).sort_values(by='StockCode')
```
