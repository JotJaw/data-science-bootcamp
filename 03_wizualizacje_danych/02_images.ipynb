```python
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```python
fig = plt.figure(figsize=(12, 10))

#enumerate - funkcja zastosowana na liście, pozwala na zastosowanie indeksu
for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(220+ idx + 1) #dwa wiersze, 2 obrazy
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
#
```

```python
fig = plt.figure(figsize=(12, 10))

#enumerate - funkcja zastosowana na liście, pozwala na zastosowanie indeksu
for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140+ idx + 1) #jeden wiersz, 4 obrazy
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
```
